## <u> Flowcharts </u>
- Flowcharts: diagrams that lay out the steps of an algorithm or process 
    - Helps to visualize the structure of the program (using technical or non-technical language 
    - Represents the specific sequence of operations 
- Specific shapes are used for different operations in a flowchart 
    - Start/stop = oval/ellipse 
    - Input/output = parallelogram 
    - Process = rectangle 
    - Decision = diamond
    
 
- Processes: this is referring to mathematical operations and changes to our information. This could be applying a formula, or multiplying/dividing/adding/etc 
- Decisions - these are parts of our flowchart where we may need different outcomes. This is when we have branching in our flowchart (a decision being made.) This depends on the logic that dictates this step, such as evaluating an inequality

- Flowcharts can be linear or non-linear depending on if they have branching paths
